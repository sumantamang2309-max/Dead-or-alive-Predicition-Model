```python
from sklearn.metrics import accuracy_score
from sklearn.linear_model import LogisticRegression
for i in range(80,500,50):
  log_model=LogisticRegression(max_iter=i)
  log_model.fit(x_train, y_train)
  log_train_acc=accuracy_score(y_train, log_model.predict(x_train))
  log_test_acc=accuracy_score(y_test, log_model.predict(x_test))
  print(f"{i}: train accuracy of log model: {log_train_acc} , and test accu is : {log_test_acc}")
```

```python
from sklearn.tree import DecisionTreeClassifier
dt_model=DecisionTreeClassifier(max_depth=3,random_state=42)
dt_model.fit(x_train,y_train)
```

# Predict and Score

```python
from sklearn.metrics import accuracy_score
log_train_acc=accuracy_score(y_train, log_model.predict(x_train))
log_test_acc=accuracy_score(y_test, log_model.predict(x_test))

dt_train_acc=accuracy_score(y_train, dt_model.predict(x_train))
dt_test_acc=accuracy_score(y_test, dt_model.predict(x_test))
```

```python
print(f"train accuracy of log model: {log_train_acc} , and test accu is : {log_test_acc}")
print(f"train accuracy of Desicion Tree model: {dt_train_acc} , and test accu is : {dt_test_acc}")
```

```python
for depth in range(2,11):
  model = DecisionTreeClassifier(max_depth=depth)
  model.fit(x_train, y_train)
  acc=accuracy_score(y_test, model.predict(x_test))
  print(f"{depth}: {acc:.3f}")
```

```python
from sklearn .ensemble import RandomForestClassifier
rf_model=RandomForestClassifier(random_state=42)
rf_model.fit(x_train,y_train)
```

```python
y_predrf=rf_model.predict(x_test)
```

```python
train_acc=rf_model.score(x_train, y_train)
test_acc=accuracy_score(y_test,y_predrf )
print(train_acc)
print(test_acc)
```

# Confusoin matrix and report

```python
cm_rf= confusion_matrix(y_test, y_predrf)
print("confusion matrux")
print(cm_rf)
print(classification_report(y_test, y_predrf))
```

```python
#Cross validation
from sklearn.model_selection import cross_val_score
scores= cross_val_score(RandomForestClassifier(), x_train, y_train, cv=5)
print("cv_score:", scores)
print("Average cv Score:", scores.mean())
```

# Grid Search : Hyperparameter

```python
from sklearn.model_selection import GridSearchCV
param_grid={"n_estimators": [50,100,150,200],
            "max_depth":["None", 3, 5 , 10 , 12],
            "min_samples_split":[2,4,6,8]

}

grid=GridSearchCV(RandomForestClassifier(),  param_grid, cv=5)
grid.fit(x_train, y_train)
print("Best parameters:", grid.best_params_)
print("Best score:", grid.best_score_)
```

```python
best_model= grid.best_estimator_
y_pred_best= best_model.predict(x_test)
```
